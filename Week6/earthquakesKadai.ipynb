    私は、USGS（アメリカ地質調査所）が公開している地震データを利用し、Pythonでインターネット上から自動的にデータを取得して表示するプログラムを作成しました。
    このプログラムでは、数字を入力することで、ダウンロードしたCSVファイルから最近の地震情報を指定した件数だけ一覧で確認することができます。
    データを自動的に取得できるため、人手で集めるよりも効率的に最新の情報を扱うことができます。
    データ量が多いため、検索機能があるのは便利だと感じています。


```python
number  = input('あなたの知りたい情報を教えてください！')
number  = int (number)

import pandas as pd
import requests

# USGSから過去1か月の地震データを取得
url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv"
response = requests.get(url)

# データをファイルに保存
with open('earthquakes.csv', 'wb') as file:
    file.write(response.content)

# データをDataFrameに読み込む
data = pd.read_csv('earthquakes.csv')

# データの最初の5行を表示
data.head(number)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV ファイル読み込み
df = pd.read_csv("earthquakes.csv")

# 条件をユーザーから取得
min_mag = float(input("最低マグニチュードを入力してください："))

# 条件に合う地震だけ抽出
cond = df["mag"] >= min_mag
filtered = df[cond]

print(f"条件に合うデータ数：{len(filtered)} 件")

# 少し確認
print(filtered[["time", "latitude", "longitude", "mag"]].head())

plt.figure(figsize=(8,5))

plt.scatter(
    filtered["longitude"],
    filtered["latitude"],
    s=filtered["mag"] * 5,
    alpha=0.6
)

plt.title("Filtered Earthquakes (Map)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```
